```python
import numpy as np
data_path='save/1D_partitioned_ssdu/531save/results.npz'
data_dict = np.load(data_path)
lst = data_dict.files
content=[]
for item in lst:
    print(item)
    content.append(item)
    # print(data_dict[item])
print(data_dict[content[0]].shape)
print(data_dict[content[1]].shape)
print(data_dict[content[2]].shape)
print(data_dict[content[3]].shape)
```

```python
import matplotlib.pyplot as plt
def plot_loss(loss,name):
    plt.figure()
    plt.plot(loss)
    plt.title(name)
    plt.show()
    plt.close('all')
```

```python
plot_loss(data_dict[content[0]],f'{content[0]} with {np.mean(data_dict[content[0]])}')
plot_loss(data_dict[content[1]],f'{content[1]} with {np.mean(data_dict[content[1]])}')
plot_loss(data_dict[content[2]],f'{content[2]} with {np.mean(data_dict[content[2]])}')
plot_loss(data_dict[content[3]],f'{content[3]} with {np.mean(data_dict[content[3]])}')
# plot_loss(data_dict[content[1]],content[1])
# plot_loss(data_dict[content[2]],content[2])
# plot_loss(data_dict[content[3]],content[3])
```
